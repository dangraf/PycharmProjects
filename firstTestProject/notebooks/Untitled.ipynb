```python
import pandas as pd
import numpy as np
import json
```

```python
df = pd.read_csv('BTC_AMP.csv')
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])

df.set_index(['date','type'], inplace=True)
df.head()
```

```python
df2 = df.groupby([pd.TimeGrouper('1min',level='date'),pd.Grouper(level='type')]).agg({'rate': np.mean, 'amount': np.sum,'total':np.sum}, inplace=True)
df2.head()
```

```python
df2.reset_index(inplace=True)
df2.head()
```
